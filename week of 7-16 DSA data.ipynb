```python
import csv 

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
df_19 = pd.read_csv("719_aspiration.csv")

df_19["width"] = (df_19["Three-phase point (r) [mm]"] - df_19["Three-phase point (l) [mm]"])/10

df_19.head()

split_loc_19 = []


for index, value in enumerate(df_19["Three-phase point (l) [mm]"]):
    if index < len(df_19) -1:
         angle_jump = (df_19.iloc[index + 1][3] - df_19.iloc[index][3])
         #print(index, angle_jump)
 
         if abs(angle_jump) > 1:
             split_loc_19.append(index)
 
print(split_loc_19)
 

# bb dye (1/10 of original 10^-5 concentration): labeled bb1

df_bb1_19_1 = df_19.iloc[:split_loc_19[0]-1, :]

df_bb1_19_2 = df_19.iloc[split_loc_19[0]:split_loc_19[1]-1, :]

df_bb1_19_3 = df_19.iloc[split_loc_19[1]:split_loc_19[2]-1, :]

df_bb1_19_4 = df_19.iloc[split_loc_19[2]:split_loc_19[3]-1, :]

df_bb1_19_5 = df_19.iloc[split_loc_19[3]:split_loc_19[4]-1, :]


df_bb1_19_6 = df_19.iloc[split_loc_19[4]:split_loc_19[5]-1, :]

df_bb1_19_7 = df_19.iloc[split_loc_19[5]:split_loc_19[6]-1, :]

df_bb1_19_8 = df_19.iloc[split_loc_19[6]:split_loc_19[7]-1, :]


#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#
#************************************************#
# bb dye (1/2 of original 10^-5 concentration) bb2
#************************************************#
#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#


df_bb2_19_1 = df_19.iloc[split_loc_19[7]:split_loc_19[8]-1, :]

df_bb2_19_2 = df_19.iloc[split_loc_19[8]:split_loc_19[9]-1, :]


df_bb2_19_3 = df_19.iloc[split_loc_19[9]:split_loc_19[10]-1, :]

df_bb2_19_4 = df_19.iloc[split_loc_19[10]:split_loc_19[11]-1, :]

df_bb2_19_5 = df_19.iloc[split_loc_19[11]:split_loc_19[12]-1, :]

df_bb2_19_6 = df_19.iloc[split_loc_19[12]:split_loc_19[13]-1, :]

df_bb2_19_7 = df_19.iloc[split_loc_19[13]:split_loc_19[14]-1, :]

df_bb2_19_8 = df_19.iloc[split_loc_19[14]:, :]




#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#
#************************************************#
# bb dye (1/10 of original 10^-5 concentration) bb1
#************************************************#
#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#

y_width1_bb1_19 = df_bb1_19_1["width"]
x_degrees1_bb1_19 = df_bb1_19_1["CA(m) []"]
plt.plot(x_degrees1_bb1_19, y_width1_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("1 s, bb dye 1/10")
plt.show()

y_width2_bb1_19 = df_bb1_19_2["width"]
x_degrees2_bb1_19 = df_bb1_19_2["CA(m) []"]
plt.plot(x_degrees2_bb1_19, y_width2_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("10 s, bb dye 1/10")
plt.show()

y_width3_bb1_19 = df_bb1_19_3["width"]
x_degrees3_bb1_19 = df_bb1_19_3["CA(m) []"]
plt.plot(x_degrees3_bb1_19, y_width3_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("20 s, bb dye 1/10")
plt.show()

y_width4_bb1_19 = df_bb1_19_4["width"]
x_degrees4_bb1_19 = df_bb1_19_4["CA(m) []"]
plt.plot(x_degrees4_bb1_19, y_width4_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("40 s, bb dye 1/10")
plt.show()

y_width5_bb1_19 = df_bb1_19_5["width"]
x_degrees5_bb1_19 = df_bb1_19_5["CA(m) []"]
plt.plot(x_degrees5_bb1_19, y_width5_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("60 s, bb dye 1/10")
plt.show()

y_width6_bb1_19 = df_bb1_19_6["width"]
x_degrees6_bb1_19 = df_bb1_19_6["CA(m) []"]
plt.plot(x_degrees6_bb1_19, y_width6_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("120 s, bb dye 1/10")
plt.show()

y_width7_bb1_19 = df_bb1_19_7["width"]
x_degrees7_bb1_19 = df_bb1_19_7["CA(m) []"]
plt.plot(x_degrees7_bb1_19, y_width7_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("180 s, bb dye 1/10")
plt.show()

y_width8_bb1_19 = df_bb1_19_8["width"]
x_degrees8_bb1_19 = df_bb1_19_8["CA(m) []"]
plt.plot(x_degrees8_bb1_19, y_width8_bb1_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("240 s, bb dye 1/10")
plt.show()


#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#
#************************************************#
# bb dye (1/2 of original 10^-5 concentration) bb2
#************************************************#
#xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx#

y_width1_bb2_19 = df_bb2_19_1["width"]
x_degrees1_bb2_19 = df_bb2_19_1["CA(m) []"]
plt.plot(x_degrees1_bb2_19, y_width1_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("1 s, bb dye 1/2")
plt.show()

y_width2_bb2_19 = df_bb2_19_2["width"]
x_degrees2_bb2_19 = df_bb2_19_2["CA(m) []"]
plt.plot(x_degrees2_bb2_19, y_width2_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("10 s, bb dye 1/2")
plt.show()

y_width3_bb2_19 = df_bb2_19_3["width"]
x_degrees3_bb2_19 = df_bb2_19_3["CA(m) []"]
plt.plot(x_degrees3_bb2_19, y_width3_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("20 s, bb dye 1/2")
plt.show()

y_width4_bb2_19 = df_bb2_19_4["width"]
x_degrees4_bb2_19 = df_bb2_19_4["CA(m) []"]
plt.plot(x_degrees4_bb2_19, y_width4_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("40 s, bb dye 1/2")
plt.show()

y_width5_bb2_19 = df_bb2_19_5["width"]
x_degrees5_bb2_19 = df_bb2_19_5["CA(m) []"]
plt.plot(x_degrees5_bb2_19, y_width5_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("60 s, bb dye 1/2")
plt.show()
    
    
y_width6_bb2_19 = df_bb2_19_6["width"]
x_degrees6_bb2_19 = df_bb2_19_6["CA(m) []"]
plt.plot(x_degrees6_bb2_19, y_width6_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("120 s, bb dye 1/2")
plt.show()

y_width7_bb2_19 = df_bb2_19_7["width"]
x_degrees7_bb2_19 = df_bb2_19_7["CA(m) []"]
plt.plot(x_degrees7_bb2_19, y_width7_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("180 s, bb dye 1/2")
plt.show()

y_width8_bb2_19 = df_bb2_19_8["width"]
x_degrees8_bb2_19 = df_bb2_19_8["CA(m) []"]
plt.plot(x_degrees8_bb2_19, y_width8_bb2_19, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("240 s, bb dye 1/2")
plt.show()

```
