```python
import csv 

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
# 10^-5 , 5x10^-6 , 10^-6
```

```python
df_88 = pd.read_csv("88_aspiration.csv")

df_88["width"] = (df_88["Three-phase point (r) [mm]"] - df_88["Three-phase point (l) [mm]"])/10

df_88.head()

split_loc_88 = []


for index, value in enumerate(df_88["Three-phase point (l) [mm]"]):
    if index < len(df_88) -1:
         angle_jump = (df_88.iloc[index + 1][3] - df_88.iloc[index][3])
         #print(index, angle_jump)
 
         if abs(angle_jump) > 1:
             split_loc_88.append(index)
 
print(split_loc_88)
 
```

# 10^-6

```python
#wait times: 1s, 10s, 40s, 60s, 120s
```

```python
df_bb1_88_1 = df_88.iloc[:split_loc_88[0]-1, :]

df_bb1_88_2 = df_88.iloc[split_loc_88[0]+1:split_loc_88[1]-1, :]

df_bb1_88_3 = df_88.iloc[split_loc_88[1]+1:split_loc_88[2]-1, :]

df_bb1_88_4 = df_88.iloc[split_loc_88[2]+1:split_loc_88[3]-1, :]

df_bb1_88_5 = df_88.iloc[split_loc_88[3]+1:split_loc_88[4]-1, :]
```

```python
# 1

y_width1_bb1_88 = df_bb1_88_1["width"]
x_degrees1_bb1_88 = df_bb1_88_1["CA(m) [°]"]
plt.plot(x_degrees1_bb1_88, y_width1_bb1_88, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("1 s, 10^-6")
plt.show()

#2

y_width2_bb1_88 = df_bb1_88_2["width"]
x_degrees2_bb1_88 = df_bb1_88_2["CA(m) [°]"]
plt.plot(x_degrees2_bb1_88, y_width2_bb1_88, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("10 s, 10^-6")
plt.show()

#3

y_width3_bb1_88 = df_bb1_88_3["width"]
x_degrees3_bb1_88 = df_bb1_88_3["CA(m) [°]"]
plt.plot(x_degrees3_bb1_88, y_width3_bb1_88, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("40 s, 10^-6")
plt.show()

#4

y_width4_bb1_88 = df_bb1_88_4["width"]
x_degrees4_bb1_88 = df_bb1_88_4["CA(m) [°]"]
plt.plot(x_degrees4_bb1_88, y_width4_bb1_88, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("60 s, 10^-6")
plt.show()

#5

y_width5_bb1_88 = df_bb1_88_5["width"]
x_degrees5_bb1_88 = df_bb1_88_5["CA(m) [°]"]
plt.plot(x_degrees5_bb1_88, y_width5_bb1_88, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("120 s, 10^-6")
plt.show()
```

```python
import matplotlib.colors as mcolors
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

plt.figure(figsize =[17,10])

colorList = cm.get_cmap("viridis", 10) 


plt.plot(x_degrees1_bb1_88, y_width1_bb1_88, 'o', label="1s", color=colorList.colors[0], markersize=7)

plt.plot(x_degrees2_bb1_88, y_width2_bb1_88, 'x', label="10s", color=colorList.colors[2], markersize=10)

plt.plot(x_degrees3_bb1_88, y_width3_bb1_88, 's', label="40s", color=colorList.colors[4], markersize=7)

plt.plot(x_degrees4_bb1_88, y_width4_bb1_88, '*', label="60s", color=colorList.colors[6], markersize=10)

plt.plot(x_degrees5_bb1_88, y_width5_bb1_88, '^', label="120s", color=colorList.colors[8], markersize=7)




plt.xlabel("angle (degrees)", fontsize=25)
plt.ylabel("width (mm)", fontsize=25)
#plt.title("brilliant blue 10^-6", fontsize=30)
plt.legend(fontsize = 40)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)

plt.xlim(0,80)
plt.ylim(0,5)

plt.show()
```
