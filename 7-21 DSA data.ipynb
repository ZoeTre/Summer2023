```python
import csv 

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt
```

```python
df_21 = pd.read_csv("721_aspiration.csv")

df_21["width"] = (df_21["Three-phase point (r) [mm]"] - df_21["Three-phase point (l) [mm]"])/10

df_21.head()

split_loc_21 = []


for index, value in enumerate(df_21["Three-phase point (l) [mm]"]):
    if index < len(df_21) -1:
         angle_jump = (df_21.iloc[index + 1][3] - df_21.iloc[index][3])
         #print(index, angle_jump)
 
         if abs(angle_jump) > 1:
             split_loc_21.append(index)
 
print(split_loc_21)
 
```

# bb 10^-6 concentration

```python

df_bb1_21_1 = df_21.iloc[:split_loc_21[0]-1, :]

df_bb1_21_2 = df_21.iloc[split_loc_21[0]:split_loc_21[1]-1, :]

df_bb1_21_3 = df_21.iloc[split_loc_21[1]:split_loc_21[2]-1, :]

df_bb1_21_4 = df_21.iloc[split_loc_21[2]:split_loc_21[3]-1, :]

df_bb1_21_5 = df_21.iloc[split_loc_21[3]:split_loc_21[4]-1, :]

df_bb1_21_6 = df_21.iloc[split_loc_21[4]+1:split_loc_21[5]-1, :]
```

```python
#all wait times
```

```python
#1

y_width1_bb1_21 = df_bb1_21_1["width"]
x_degrees1_bb1_21 = df_bb1_21_1["CA(m) [°]"]
plt.plot(x_degrees1_bb1_21, y_width1_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("1 s, bb dye 10^-6")
plt.show()

#2

y_width2_bb1_21 = df_bb1_21_2["width"]
x_degrees2_bb1_21 = df_bb1_21_2["CA(m) [°]"]
plt.plot(x_degrees2_bb1_21, y_width2_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("10 s, bb dye 10^-6")
plt.show()

#3

y_width3_bb1_21 = df_bb1_21_3["width"]
x_degrees3_bb1_21 = df_bb1_21_3["CA(m) [°]"]
plt.plot(x_degrees3_bb1_21, y_width3_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("20 s, bb dye 10^-6")
plt.show()

#4

y_width4_bb1_21 = df_bb1_21_4["width"]
x_degrees4_bb1_21 = df_bb1_21_4["CA(m) [°]"]
plt.plot(x_degrees4_bb1_21, y_width4_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("40 s, bb dye 10^-6")
plt.show()

#5

y_width5_bb1_21 = df_bb1_21_5["width"]
x_degrees5_bb1_21 = df_bb1_21_5["CA(m) [°]"]
plt.plot(x_degrees5_bb1_21, y_width5_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("60 s, bb dye 10^-6")
plt.show()

#6

y_width6_bb1_21 = df_bb1_21_6["width"]
x_degrees6_bb1_21 = df_bb1_21_6["CA(m) [°]"]
plt.plot(x_degrees6_bb1_21, y_width6_bb1_21, 'o')
plt.xlabel("angle (degrees)")
plt.ylabel("width (mm)")
plt.title("120 s, bb dye 10^-6")
plt.show()
```
